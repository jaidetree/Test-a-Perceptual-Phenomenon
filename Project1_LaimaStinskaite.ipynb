**Question 1**:  What is our independent variable? What is our dependent variable?

**Answer**: 
1. Independent variable is "two conditions: a congruent words condition, and an incongruent words condition".
2. Dependent variable is "time it takes to name the ink colors in equally-sized lists".

**Question 2**: What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

**Answer**: 

Uc -- the mean of time for congruent words condition for population

Ui --  the mean of time for incongruent words condition for population

Ud = Ui - Uc -- the difference of Ui and Uc means

H0 -- null hypothesis

Ha -- alternative hypothesis

H0: Ud = 0 (there is no difference between those sets of times for the population)

Ha: Ud > 0 (it takes more time to say color of ink for incongruent words condition then for congruent words condition for whole population)

What we know\don't know:

1. We don't know mean and standard deviation of whole population
2. But we know that for congruent and incongruent sets the same people were experimented
3. And we want to see if the sample difference is statistically significant or not

So, from this information we decide that a type of our test will be an one-tailed dependent t-test for paired sample.

**Question 5**: Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

**Answer**: 

1. Mean difference d = Xi - Xc = 22.02 - 14.05 = 7.97
2. Standard deviation of the differences Sd = sqrt [ (Î£(di - d)^2 / (n - 1) ] = 4.86
3. Standard error of the mean difference SE(d) = Sd / sqrt (n) = 4.86 / sqrt(24) = 0.99
4. t-statistic t = d / SE(d) = 7.97 / 0.99 = 8.05
5. Degree of freedom DF = n - 1 = 23
6. alpha-level = 0.05
7. t-critical = 1.714
8. probability p <.0001

The one-tailed p value is less than 0.0001. By conventional criteria, this difference is considered to be extremely statistically significant (< 0.05). Therefore we reject H0. It means that it takes more time to say color of the ink from incongruent set than from congruent.
